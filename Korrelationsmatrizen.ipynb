```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Daten laden
df = pd.read_csv('C:\\Desktop\\DatensätzeCSV\\BMW20232021CSV\\BMW2021CSV.csv', delimiter=';')

# Ändere die Reihenfolge der Variablen
new_order = ['price', 'mileage', 'hp', 'year']  # Neue Reihenfolge der Variablen
df_reordered = df[new_order]

# Entferne nicht-numerische Spalten
numerical_columns = ['price', 'mileage', 'hp', 'year']
df_numerical = df[numerical_columns]

# Korrelationsmatrix berechnen
correlation_matrix = df_numerical.corr()

# Korrelationsmatrix ausgeben
print("Korrelationsmatrix:")
print(correlation_matrix)

# Korrelationsmatrix visualisieren
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Korrelationsmatrix BMW 100-200 PS, Automatik, Diesel, 2021')
plt.show()
```
